En neuroinformática se utilizan modelos para simular el funcionamiento de las neuronas. Lo más sencillo de simular son los disparos. Veamos un ejemplo de disparo:

image.png

Como vemos el voltaje de la neurona  deoende de distintos canales. En la imágen anterior solo se tienen en cuenta los de sodio y potasio. El modelo que vamos a utilizar describe la variación del voltaje como una ecuación diferencial que depende de otra ecuación diferencial que simula los canales. Veamos las ecuaciones:

\begin{equation}
\frac{dx}{dt} = y + ax^2 - bx^3 - I
\end{equation}

\begin{equation}
\frac{dy}{dt} = c - dx^2 - y
\end{equation}

Más especificamente: 

* X es el potencial de membrana debido a las diferentes concentraciones iónicas intracelulares y extracelulares de calcio, potasio y sodio a través de la membrana y se mide mediante la variable x(t).

* Y es La variable de pico, también conocida como corriente de recuperación, es el transporte de iones de sodio y potasio a través de canales rápidos activados por iones y está modelada por la variable y(t).

Estas ecuaciones diferenciales se pueden resolver numéricamente, es decir, partiendo de unas condiciones iniciales para x e y, podemos iterar un algoritmo que calcule la siguiente iteración, dando x e y. El más sencillo es Euler. Simplemente,

\begin{equation}
\frac{\vartriangle x}{\vartriangle t} =  (y_{i} + ax_{i}^2 - bx_{i}^3 - I)
\end{equation}

\begin{equation}
\vartriangle x = \vartriangle t (y_{i} + ax_{i}^2 - bx_{i}^3 - I)
\end{equation}



Donde podemos fijar $dt=\delta$ y separa $dx=x_{i+1}-x_{i}$:

\begin{equation}
x_{i+1} = x_{i} + \delta(y_{i} + ax_{i}^2 - bx_{i}^3 - I)
\end{equation}

\begin{equation}
y_{i+1} = y_i + \delta(c -dx_{i}^2 - y_i)
\end{equation}
