# **Práctica 2: Encontrar subsecuencia**



Dada una secuencia de ADN encontrar el motif (subsecuencia) indicado.

ATTCAAAATCTTTTAAAAAATTAACTAACTATA


Una forma de plantear el problema es realizando un autómata finito. Los autómatas son un conjunto de estados conectados por aristas. Veamos un ejemplo, si queremos reconocer la secuencia de ADN "ATA":


image.png

Las letras de la secuecia van a ir entregandose al autómata en el estado en el que se encuentre en ese momento. Va a avanzar hasta llegar al estado final (dos círculos). El automáta está totalmente definido, cada estado tiene conexiones para todas las posibilidades. Los automátas son muy fáciles de programar y se pueden construir de manera muy sencilla para encontar palabras cortas. Cuando las palabras son más largas se vuelve más complicado.
